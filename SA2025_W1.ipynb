```python
common_origin = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]['origin'].mode()[0]
print("\n3. Most common origin:", common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
japan_accel = round(df[df['origin'] == 'japan']['acceleration'].mean(), 2)
print("\n4. Mean acceleration (Japan):", japan_accel)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = df['mpg'].median()
best_ratio_car = df[df['mpg'] > median_mpg].assign(
    hp_weight_ratio=lambda x: x['horsepower']/x['weight']
).sort_values('hp_weight_ratio', ascending=False).iloc[0].name
print("\nUngraded: Best HP/weight ratio car:", best_ratio_car)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for origin in df['origin'].unique():
    subset = df[df['origin'] == origin]
    yearly_mpg = subset.groupby('model_year')['mpg'].mean()
    plt.plot(yearly_mpg.index, yearly_mpg.values, label=origin, marker='o')

plt.title('Average MPG Evolution by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend()
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df is already loaded and potentially has a non-unique index
# Reset the index to ensure it is unique before plotting
df_reset = df.reset_index(drop=True)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    # Use the DataFrame with a reset index
    data=df_reset,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),
    alpha=0.7
)
plt.title('Car Performance Characteristics')
plt.xlabel('Horsepower')
plt.ylabel('Weight (lbs)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
consistent_cars = (df.groupby('name')
                   .filter(lambda x: x['model_year'].nunique() > 1)
                   .groupby('name')
                   .agg(
                       appearances=('model_year', 'count'),
                       mpg_std=('mpg', 'std'),
                       avg_mpg=('mpg', 'mean')
                   )
                   .query('mpg_std < 1.0 and appearances >= 2')
                   .sort_values(['appearances', 'avg_mpg'], ascending=[False, False])
                  )
print("\nConsistent cars:")
print(consistent_cars)
```
